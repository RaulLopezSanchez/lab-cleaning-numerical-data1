```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python

print(customer_df.head())


print(customer_df.shape)


print(customer_df.info())
```

```python
# Renombrar las columnas utilizando snake case
customer_df.columns = customer_df.columns.str.lower().str.replace(' ', '_')
```

```python
# Cambiar el tipo de datos de cualquier otra columna seg√∫n sea necesario
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
```

```python
# Calcular la matriz de correlaci√≥n
correlation_matrix = customer_df.corr()

# Graficar la matriz de correlaci√≥n utilizando un mapa de calor
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de correlaci√≥n')
plt.show()
```
